Let us do three clusters

### The silhouette coefficient

```python
from sklearn.metrics import silhouette_score
silhouette_score(song_two_norm_df, song_two_norm_df.cluster)
```

The silhouette coefficient can vary between -1 and +1:

|   |   |
|---|---|
|+1   | the instance is well inside its own cluster and far from other clusters.  |
|0   | the instance is close to a cluster boundary.   |
|-1   | the instance may have been assigned to the wrong cluster.  |

```python
sil_scores = []
for i in range(2,15):
  my_kmeans = KMeans(n_clusters=i)
  my_kmeans.fit(song_two_norm_df)
  song_two_norm_df['cluster'] = my_kmeans.labels_
  sil_scores.append(silhouette_score(song_two_norm_df, song_two_norm_df.cluster))

df_sc_mean=pd.DataFrame(sil_scores)
```

```python
df_sc_mean.plot();
```

### Inertia and the Elbow Method

```python
list_k = []
seed = 0
for i in range(5,20):
  my_kmeans = KMeans(n_clusters=i, random_state=seed, n_init=10)
  my_kmeans.fit(song_1000_norm_mm_df)
  list_k.append(my_kmeans.inertia_)
df_k_mean=pd.DataFrame(list_k)
```

```python
df_k_mean.plot();
plt.annotate('Elbow', xy=(5, 600), xytext=(6, 100000), arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel("K")
plt.ylabel("Inertia");
```

```python
max_k = 20
inertia = []

for k in range(5, max_k):
    inertia.append(KMeans(n_clusters=k).fit(song_1000_norm_mm_df).inertia_)

fig, ax = plt.subplots(figsize=(12, 6))
plt.title(f'Inertia evolution from 5 cluster to {max_k} cluster')
sns.lineplot(x=range(5, max_k), y=inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

## t-SNE

#### With 8 clusters

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import seaborn as sns

km = KMeans(n_clusters=8,random_state=seed, n_init=10)
km.fit(song_1000_norm_mm_df)

#Add the cluster labels to the original dataframe
song_1000_norm_mm_df['Cluster'] = km.labels_

#Perform t-SNE to reduce dimensionality to 2 components
tsne = TSNE(n_components=2, random_state=0)
tsne_components = tsne.fit_transform(song_1000_norm_mm_df.drop('Cluster', axis=1))

#Create a new DataFrame with the t-SNE components
tsne_df = pd.DataFrame(data=tsne_components, columns=['Component 1', 'Component 2'])

#Add the cluster labels to the t-SNE dataframe
tsne_df['Cluster'] = km.labels_

#Define a custom color palette for each cluster
#custom_palette = sns.color_palette("Set1", n_colors=km.n_clusters)

#Create a scatter plot of the t-SNE components with custom colors
sns.set(style="ticks")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tsne_df, x='Component 1', y='Component 2', hue='Cluster', markers=["o", "s", "D", "P"])
plt.title('t-SNE Scatter Plot of Clusters')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend(title='Cluster')
plt.show()
```
